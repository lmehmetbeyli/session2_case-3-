## Basic matplotlib exercises

#### Import matplotlib and numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Identify x=[0,6,7],y=[9,10,11], give names to x ;y axes, to the label and title , show legend

```python
fig = plt.figure()
x=[0,6,7]
y=[9,10,11]
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('X and Y')
fig
```

####  Control colours ;

The colour names and colour abbreviations are given in the following table:-


**Colour abbreviation**      **Colour name**

b                               blue

c                               cyan

g                               green

k                               black

m                               magenta

r                               red

w                               white

y                               yellow

There are several ways to specify colours, other than by colour abbreviations:
    
•	The full colour name, such as yellow

•	Hexadecimal string such as ##FF00FF

•	RGB tuples, for example (1, 0, 1)

•	Grayscale intensity, in string format such as ‘0.7’.


```python
x16 = np.arange(1, 5)
# Plot green line with x16
plt.plot(x16,'g')

# Plot red line with x16+1
plt.plot(x16+1,'r')

# Plot blue line with x16+2
plt.plot(x16+2,'b')


```

## Control line styles

All the available line styles are available in the following table:


**Style abbreviation**  **Style**

-                    solid line
   
--                   dashed line
   
-.                   dash-dot line
   
:                    dotted line
   


Now, we can see the default format string for a single line plot is 'b-'.

```python
x16 = np.arange(1, 5)
# plot blue line with -- , red line with -. , green line with :
plt.plot(x16, '--', x16+1, '-.', x16+2, ':')
```

```python
x = np.linspace(0, 2)
plt.plot(x,x, label='linear')
plt.plot(x,x**2, label='quadratic')
plt.plot(x,x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()

# plot x and x and label it as 'linear'
# plot x and x**2 and label it as 'quadratic'
# plot x and x**3 and label it as cubic
# label x and y axes
# show legend
```
